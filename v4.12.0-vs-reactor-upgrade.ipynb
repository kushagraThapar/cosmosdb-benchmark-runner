```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

%matplotlib inline
```

# Point Read Throughput on Standard D16s v3 (16 vcpus, 64 GiB memory)

```python
f2_pointread = pd.read_csv('benchmark-results-point-read-4.12.0.csv', skipinitialspace=True)
sns.lineplot(x='ThreadCount', y ='Throughput(ops/s)', hue='OpName', data=(f2_pointread[f2_pointread.ThreadCount < 300]))
```

# Point Read Latency on Standard D16s v3 (16 vcpus, 64 GiB memory) 

```python
sns.lineplot(x='ThreadCount', y ='P95(ms)', hue='OpName', data=(f2_pointread[(f2_pointread.ThreadCount < 100) & (f2_pointread.ThreadCount > 0)]))
```
