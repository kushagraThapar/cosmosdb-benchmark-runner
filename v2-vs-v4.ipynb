```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

%matplotlib inline
```

# This notebook visualizes the perf of sync v2.6.1 and v4.11.0 SDKs.


## PointRead on 2 CPU cores VM (F2s)

```python
f2_pointread = pd.read_csv('notebooks/20210119/f2-point-read.csv', skipinitialspace=True)
sns.lineplot(x='ThreadCount', y ='Throughput(ops/s)', hue='OpName', data=(f2_pointread[f2_pointread.ThreadCount < 100]))
```

```python
sns.lineplot(x='ThreadCount', y ='P95(ms)', hue='OpName', data=(f2_pointread[(f2_pointread.ThreadCount < 100) & (f2_pointread.ThreadCount > 0)]))
```

## Query vs ReadMany on 2 CPU cores VM (F2s)

```python
sns.lineplot(x='ThreadCount', y ='P95(ms)', hue='OpName', data=(f2_query[(f2_query.ThreadCount < 100) & (f2_query.ThreadCount > 0)]))
```
